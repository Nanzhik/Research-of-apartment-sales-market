## Предобработка данных

Прдеварительную обработку данных будем производить постепенно по каждому столбцу, группируя обработку столбцов по их смысловому значению.

Для более быстрого предоставления данных по каждому количественному столбцу напишем функцию, раскрывающую детали, а также функцию, строющую гистограмму:

```python
def column_describe(column):
    '''
    Считает кол-во пропусков в столбце,
    Определяет тип данных значений в столбце,
    Выводит детальное описание столбца
    '''
    fills = data[column].isna().sum()
    d_type = data.dtypes[column]
    describe = data[column].describe()
    print('Название столбца:', column)
    print('Количество пропусков:', fills)
    print('Тип данных:', d_type)
    print('Детальное описание:')
    print(describe)
    
def hist_func(column, x_name, title_name):
    '''
    Строит гистограмму по указанному столбцу,
    Подписывает ось Х и название гистограммы
    '''
    plt.figure(figsize=(10,5))
    plt.hist(data[column])
    plt.grid()
    plt.xlabel(x_name)
    plt.title(title_name)
    plt.show()
```

### Переименование столбцов

### Этапы предобработки данных

1. *Изучение пропусков*  
Ранее определили, что во многих столбцах имеются пропуски. Заполним пропуски там, где это возможно.  

2. *Рассмотрение типов данных*  
Было определено, что типы данных некоторых столбцов не соответствуют их значениям. Будем менять типы данных на необходимые одновременно с заполнением пропусков для сокращения строк кода.

3. *Устранение неявных дубликатов*  
Изучим уникальные значения в столбцах с названиями и устраним неявные дубликаты.

4. *Устранение редких и выбивающихся значений*  
Будем восстанавливать корректные значения в случае наличия аномалий там, где это возможно. В остальных случаях будем удалять редкие и выбивающиеся значения.

### `'balcony'` 

Выведем информацию о столбце при помощи функции column_describe

```python
column_describe('balcony')
```

Там, где не указано число балконов поставим 0, т.к. скорее всего где балконы не указаны, значит их нет. 
По полученной информации у данного столбца указан тип данных с плавающей точкой - заменяем на целочисленный (Т.к. балкон либо есть, либо его нет. Не может быть десятичной доли балкона).

```python
data['balcony'] = data['balcony'].fillna(value=0).astype('int')

# Проверим кол-во пропущенных значений в столбце 'balcony'
data['balcony'].isna().sum()
```

```python
# построим гистрограмму распределения балконов по объявлениям
hist_func('balcony',
          'Количество балконов',
          'Распределение балконов по объявлениям'
)
```

### `'is_apartment'`

Выведем информацию о столбце при помощи функции column_describe

```python
column_describe('is_apartment')
```

Пропуски в апартаментах заменим нулями - будем считать, что помещения с неуказанным данным полем не являются апартаментами.
По полученной информации увидели, что тип данных данного столбца 'object' - заменяем на булевый (помещение либо является апартаментом - True, либо нет - False).

```python
def mean_count(area, total_area):
    for t in data[total_area].unique():
        data.loc[(data[total_area] == t) & (data[area].isna()), area] = \
        data.loc[(data[total_area] == t), area].mean()
        
    return print('Количество оставшихся пропущенных значений в столбце', 
                 area,
                 ':',
                 data[area].isna().sum())

mean_count('living_area', 'total_area')
mean_count('kitchen_area', 'total_area')
```

Теперь остались лишь около 1% пропусков от общего числа значений. С данными проупсками ничего предпринимать не будем.

Будем считать, что произошла ошибка при занесении данных и такие значения будем делить на 10. Квартиры, высотой меньше 25 пока оставляем как есть (запомним, вернемся к ним позднее), потому что не может быть квартира высотой менее 2,5 метров. 

```python
data['ceiling_height'] = data['ceiling_height'].replace(25, 2.5)
data['ceiling_height'] = data['ceiling_height'].replace(26, 2.6)
data['ceiling_height'] = data['ceiling_height'].replace(27, 2.7)
data['ceiling_height'] = data['ceiling_height'].replace(32, 3.2)
```

Теперь рассмотрим квартиры, высота потолков в которых меньше 2,5 м и больше 5 метров. Данные высоты являются аномальными значениями. Посчитаем общее кол-во таких квартир.

```python
data[data['ceiling_height'] < 2.5]['ceiling_height'].count() + data[data['ceiling_height'] > 5]['ceiling_height'].count()
```

Предлагается следующий способ обработки таких значений:  
1. Сначала обнулим эти значения
2. Затем вместе с оставшимися пропущенными заполним пропуски медианными значениями, в зависимости от цены квартиры и категорией квартиры (апартаменты или нет).  

Если бы в датасете были данные о годе постройки дома, то можно было бы заполнить пропуски в зависимости от этажности и года постройки. Но так как данная информация отсутствует, то предлагается разбить квартиры в зависимости от цены на квартиры эконом-класса и бизнес-класса, если квартира не является апартаментами. Апартаменты вынесем в отдельную категорию. 

Границу для определения цены квартиры бизнес-класса определим из ранее полученных данных столбца `'last_price'`. 

```python
q25_price = data.last_price.quantile(0.25)
q75_price = data.last_price.quantile(0.75)
iqr_price = q75_price - q25_price
upper_price = q75_price + 1.5 * iqr_price
lower_price = q25_price - 1.5 * iqr_price

q25_price, q75_price, iqr_price, upper_price, lower_price
```

Согласно посчитанном разбросу цен, примем, что бизнес-классом будут считаться квартиры, цена которых начинается от верхней границы (11.8 млн.руб.)

Создадим вспомогательный столбец, в который определим категорию жилья, в зависимости от цены.

```python
def price_category(row):
    '''
    Определяет категорию жилью в зависимости от стоимости на квартиру
    и статусом квартиры (является помещение апартаментами или нет)
    '''
    is_apart = row['is_apartment']
    price = row['last_price']
    
    if ~is_apart and price < 11800000:
        return 'econom'
    elif ~is_apart and price >= 11800000:
        return 'business'
    return 'apartment'

# Добавим новый столбец 
data['price_category'] = data.apply(price_category, axis=1)
# Просмотрим первые 5 строк таблицы
data.head()
```

```python
# Обнулим аномальные значения высот квартир
data.loc[(data['ceiling_height'] < 2.5) | (data['ceiling_height'] > 5), 'ceiling_height'] = None

# Заполним все пропуски медианными значениями в зависимости от категории
for p in data['price_category']:
    data.loc[(data['price_category'] == p) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
    data.loc[(data['price_category'] == p), 'ceiling_height'].median() 

# Посмотрим обновленное распределение квартир
column_describe('ceiling_height')
hist_func(
    'ceiling_height', 
    'Высота потолков', 
    'Распределение высоты потолков квартир в объявлениях'
)
```

Пропуски высоты потолков заполнены. Наибольшее кол-во квартир высотой 2,5-2,7 м.
Вспомогательный столбец с категорией стоимости пока оставим, может понадобиться при дальнейшем анализе.

Изучим наиболее частые высоты потолков в квартирах подробнее

```python
plt.figure(figsize=(10,5))
plt.hist(data['ceiling_height'], bins=10, range=(2.5, 3.5))
plt.grid()
plt.xlabel('Высота потолков, м')
plt.title('Распределение высоты потолков квартир в объявлениях')
plt.show()
```

Прошлые выводы подтверждаются: наибольшее кол-во квартир высотой 2,5-2,7 м.

## Исследовательский анализ данных

Отдельные параметры объектов описаны и изучены ранее. Изучим факторы, влияющие на общую (полную) стоимость объекта.

```python
# Создадим дополнительную переменную с необходимы столбцами для сравнения
data_corr = data[['last_price', 
                  'total_area', 
                  'living_area',
                  'kitchen_area',
                  'rooms', 
                  'type_floor', 
                  'weekday_exposition', 
                  'month_exposition', 
                  'year_exposition']]
```

Построим диаграммы размаха отдельно для каждого параметра.

```python
def boxplot_builder(column, ylim_max):
    '''
    Строит вспомогательную таблицу и диаграмму размаха 
    стоимости на жилье в зависимости от указанного столбца
    '''
    data_pivot = data.pivot_table(
        index=column,
        values='last_price',
        aggfunc='mean'
    ).fillna(0)
    print(data_pivot)
    data.boxplot('last_price', by=column, figsize=(10, 10)).set_ylim(0, ylim_max)
```

```python
boxplot_builder('weekday_exposition', 10)
```

Как и предполагалось выше, от дня недели размещения объявления стоимость жилья не зависит.

```python
boxplot_builder('month_exposition', 10)
```

Аналогично предыдущему графику, от месяца размещения объявления стоимость жилья не зависит.

```python
boxplot_builder('year_exposition', 15)
```

С годом размещения ситуация немного другая: в 2014 году был скачок цен, далее цены держались плюс-минус на одном уровне.

```python
boxplot_builder('rooms', 50)
```

От количества комнат связь стоимости жилья практически прямая - чем больше комнат, тем дороже жилье, что логично.

```python
def scatter_builder(column):
    '''
    Строит вспомогательную таблицу и график корреляции 
    цены на квартиру в зависимости от указанного столбца
    '''
    data_pivot = data.pivot_table(
        index=column,
        values='last_price',
        aggfunc='mean'
    ).fillna(0)
    print(data_pivot)
    data.plot(x=column, y='last_price', kind='scatter', grid=True, alpha=0.3)
```

```python
scatter_builder('total_area')
```

Зависимость стоимости от общей площади практически прямая - чем больше площадь квартиры, тем дороже жилье. Все логично.
